```python
import os
import glob
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
```

```python
hdf_files = glob.glob('../data/features_03s/*.h5')
```

```python
dfs = []
for hdf_path in hdf_files:
    name = os.path.splitext(os.path.basename(hdf_path))[0]
    df = pd.read_hdf(hdf_path)
    df.insert(0, 'name', name)
    dfs.append(df)
    
dfs = pd.concat(dfs)
print(f'Found {len(dfs)} records')
dfs.head(5)
```
